# Marketing data analytical metrics

## Initial Metrics
1. Budget
2. Goal: click, sales, call

## Business Performance Metrics
1. **CPA - Cost Per Acquisition:** The cost to acquire a lead or customer. Indicates the efficiency of your campaigns in generating valuable leads. Test ad channels, refine targeting, and focus on high-value conversions.
1. **ROI - Return on Investment:** The revenue generated from leads compared to the cost of acquiring them. Reflects the overall profitability of your lead generation campaigns.
1. **LTV - Lifetime Customer Value:** Average of total amount of spent of customers
1. **LVR - Lead Velocity Rate:** The growth rate of qualified leads month-over-month. Indicates whether your lead generation strategies are scaling effectively.

## Lead Metrics
1. **CLV - Customer Lifetime Value:** The estimated total value a lead brings over time. Helps determine if lead acquisition costs are justified based on long-term revenue.
1. **Lead-to-Conversion Rate:** The percentage of leads that convert into paying customers. Directly reflects lead quality and overall campaign success. Focus on targeting high-intent audiences and refining your lead nurturing process.
1. **CPQL - Cost Per Qualified Lead:** The cost of acquiring a lead that meets qualification criteria. Focuses on the efficiency of generating valuable leads rather than overall volume. Reduce spend on low-performing channels or audiences.
1. **Lead scoring:** A scoring system that evaluates leads based on behavior, demographics, and firmographics. Prioritizes leads most likely to convert. Continuously refine the scoring model based on conversion data.
1. **Lead Engagement Metrics:** The degree of interaction with your brand, e.g., email open rates. Indicates interest and readiness to engage further. Create engaging follow-ups and nurture sequences.

## Website Metrics
1. Sessions: Total amount of time a new session has started on your website
1. Unique Visitors: How many unique visitors have visited your website
1. **Bounce Rate:** The percentage of visitors who leave your landing page without interacting. Indicate a disconnect between your ad and landing page content. Improve landing page relevance, design, and loading speed.
1. **CVR - Conversion Rate:** The percentage of ad clicks that lead to a desired action. Directly measures how well your ad drives conversions. Ensure ad targeting matches the audience intent, align landing pages with ad content, and optimize user experience.

## Ad/Social Media Metrics
1. Impression: Total amount of people who have seen your post/ads, even if they're the same person
1. CPM - Cost Per Mille: How much to get 1000 Impression
1. Reach: How many unique individual have seen that post/ads
1. Clicks/Calls/Visits: Measurement of how many people have clicked/called/visited you
1. Frequency: How often a person has seen your ads, e.g. 15 Frequency = 1 person has seen your ads 15 times
1. **CTR - Click Through Rate:** The percentage of people who clicked your ad after seeing it. Indicates that your ad creative and messaging resonate with the audience. Optimize ad copy, visuals, and call-to-action (CTA).
1. **CPC - Cost Per Click:** The cost incurred for each click on your ad. Helps evaluate ad efficiency and optimize spend for better ROI. Adjust bidding strategies, focus on high-performing keywords, or refine audience targeting.

```python
# Data Collection - 1
targetDirName = "facebook"
mainDirPath = "drive/MyDrive/data_storage/marketing_channel_data/" + targetDirName
fullDirPath = mz.getPath("content",mainDirPath)
files = os.listdir(fullDirPath)

files
```

```python
# Data collection - 2
df_raw = pd.DataFrame()

for e in files:
  df_temp = pd.read_csv(fullDirPath / e)
  df_raw = pd.concat([df_raw, df_temp], ignore_index=True)

print(df_raw.info())
print(df_raw.describe())
df_raw.head()
```

```python
# Data Cleaning - 0
##################################################
df_prep = df_raw.copy()
df_prep.head()
```

```python
# Data Cleaning - 1
##################################################
# Static feature names, so less chance of typing error
__seperator = "_"
#
__ad_id               = "ad_id"
__reporting_start     = "reporting_start"
__reporting_end       = "reporting_end"
__campaign_id         = "campaign_id"
__fb_campaign_id      = "fb_campaign_id"
__age                 = "age"
__gender              = "gender"
__interest1           = "interest1"
__interest2           = "interest2"
__interest3           = "interest3"
__impressions         = "impressions"
__clicks              = "clicks"
__spent               = "spent"
__total_conversion    = "total_conversion"
__approved_conversion = "approved_conversion"
```

```python
# Data Cleaning - 2
##################################################
# Checking some duplicates
df_temp = df_prep[df_prep.duplicated(keep=False)]
print(len(df_temp))
df_temp
```

```python
# Data Cleaning - 3.1
##################################################
# Checking some missing values
df_temp = df_prep[df_prep.isnull().any(axis=1)]
print(len(df_temp))
df_temp
```

```python
# Data Cleaning - 3.2
##################################################
# It seems like Campaign/FB Campaign IDs are missing
df_prep.dropna(inplace=True)
df_prep.head()
```

```python
# Data Cleaning - 4.1
##################################################
# Checking some ID features
id_list = [__ad_id, __campaign_id, __fb_campaign_id, __total_conversion, __approved_conversion]
for e in id_list:
  temp_list = df_prep[e].unique()
  print(e,len(temp_list))
```

```python
# Data Cleaning - 4.2
##################################################
# Changing some ID names
__campaign = "campaign"
#
temp_list = df_prep[__campaign_id].unique()
for i, id in enumerate(temp_list):
  df_prep[__campaign_id] = df_prep[__campaign_id].replace({id:__campaign+__seperator+str(i+1)})
df_prep[__campaign_id].unique()
```

```python
# Data Cleaning - 4.3
##################################################
# Checking some target values
print(df_prep[__total_conversion].unique())
print(df_prep[__approved_conversion].unique())
```

```python
# Data Cleaning - 5.1
##################################################
# Data Types
df_prep.dtypes
```

```python
# Data Cleaning - 5.2
##################################################
# Changing Data Types
date_list = [__reporting_start,__reporting_end]
for e in date_list:
  df_prep[e] = pd.to_datetime(df_prep[e])
df_prep.dtypes
```

```python
# Data Cleaning - 6
##################################################
# Index setting
index = np.arange(1,len(df_prep)+1)
df_prep.set_index(index, inplace=True)
df_prep.head()
```

```python
import matplotlib.ticker as mtick
```

```python
# EDA - Descriptive statistics - 1
##################################################
# Campaign group
_feature = __campaign_id
_groupby = __campaign_id
_target = __approved_conversion
order_list = df_prep[_feature].unique()
#
ax = sns.countplot(x=_feature, data=df_prep, palette='colorblind', legend=False, order=order_list)
total = df_prep[_feature].count()
for p in ax.patches:
    percentage = '{:.1f}%'.format(100 * p.get_height() / total)
    x = p.get_x() + p.get_width() / 2 - 0.15
    y = p.get_y() + p.get_height()
    ax.annotate(percentage, (x, y), size=12)
plt.xticks(rotation=0, ha='center')
plt.show()
#
df_temp = df_prep.groupby(_feature)[_target].sum().reset_index()
total = df_temp[_target].sum()
ax = sns.barplot(x=_feature, y=_target, palette='colorblind', data=df_temp, order=order_list)
for p in ax.patches:
    percentage = '{:.1f}%'.format(100 * p.get_height() / total)
    x = p.get_x() + p.get_width() / 2 - 0.15
    y = p.get_y() + p.get_height()
    ax.annotate(percentage, (x, y), size=12)
plt.ylabel(_target)
plt.xticks(rotation=0, ha='center')
plt.show()
```

```python
df_melted = pd.melt(df_prep, id_vars=[__campaign_id],
                    value_vars=[__approved_conversion, __total_conversion],
                    var_name='Conversion Type', value_name='Count')

total_counts = df_melted.groupby([__campaign_id, 'Conversion Type'])['Count'].sum().reset_index()

ax = sns.barplot(x=__campaign_id, y='Count', hue='Conversion Type', data=total_counts, palette='colorblind')

plt.ylabel('Conversion Count')
plt.xticks(rotation=0, ha='center')
plt.legend(title='Conversion Type')

for i, p in enumerate(ax.patches):
  if (i >= len(total_counts)):
    break
  campaign_id = df_melted[__campaign_id].unique()[i % len(df_melted[__campaign_id].unique())]
  conversion_type = df_melted['Conversion Type'].unique()[int(i / len(df_melted[__campaign_id].unique()))]
  current_count = total_counts[(total_counts[__campaign_id] == campaign_id) & (total_counts['Conversion Type'] == conversion_type)]['Count'].values[0]
  campaign_total_count = total_counts[total_counts[__campaign_id] == campaign_id]['Count'].sum()
  conversion_total_count = total_counts[total_counts['Conversion Type'] == conversion_type]['Count'].sum()

  percentage_1 = 'CAM {:.1f}%'.format(100 * current_count / campaign_total_count)
  percentage_2 = 'CVR {:.1f}%'.format(100 * current_count / conversion_total_count)
  display = percentage_1 + "\n" + percentage_2
  x = p.get_x() + p.get_width() / 2 - 0.15*1.5
  y = p.get_y() + p.get_height()
  ax.annotate(display, (x, y), size=10)
plt.show()
```

We can see some statistics about this dataset
+ Campaign 2 is over 61% of the total campaign count
+ Campaign 3, while only 31% of the total campaign, responsible for over 64% of the total Approved conversion

```python
# EDA - Descriptive statistics - 2
##################################################
# Age group
_feature = __age
_groupby = __campaign_id
_target = __approved_conversion
order_list = df_prep[_feature].unique()
#
ax = sns.countplot(x=_feature, data=df_prep, palette='colorblind', legend=False, order=order_list)
total = df_prep[_feature].count()
for p in ax.patches:
    percentage = '{:.1f}%'.format(100 * p.get_height() / total)
    x = p.get_x() + p.get_width() / 2 - 0.15
    y = p.get_y() + p.get_height()
    ax.annotate(percentage, (x, y), size=12)
plt.xticks(rotation=0, ha='center')
plt.show()
#
df_temp = df_prep.groupby(_feature)[_target].sum().reset_index()
total = df_temp[_target].sum()
ax = sns.barplot(x=_feature, y=_target, palette='colorblind', data=df_temp, order=order_list)
for p in ax.patches:
    percentage = '{:.1f}%'.format(100 * p.get_height() / total)
    x = p.get_x() + p.get_width() / 2 - 0.15
    y = p.get_y() + p.get_height()
    ax.annotate(percentage, (x, y), size=12)
plt.ylabel(_target)
plt.xticks(rotation=0, ha='center')
plt.show()
#
df_temp = df_prep.groupby(__campaign_id)[__approved_conversion].sum()
df_grouped = df_prep.groupby([__campaign_id, _feature])[__approved_conversion].sum().reset_index()
df_grouped['percentage'] = df_grouped.apply(lambda row:
                                      100 * row[__approved_conversion] / df_temp[row[__campaign_id]],
                                      axis=1)

ax = sns.barplot(x=__campaign_id, y='percentage', hue=_feature, data=df_grouped, errorbar=None, hue_order=order_list)
ax.yaxis.set_major_formatter(mtick.PercentFormatter())
plt.ylabel(__approved_conversion + " Percentage")
plt.xticks(rotation=0, ha='center')
plt.show()
```
